```python
import pandas as pd
```

```python
# データの読み込み
train_data = pd.read_csv('../data/train.csv')
test_data = pd.read_csv('../data/test.csv')
card_data = pd.read_csv('../data/card.csv')
user_data = pd.read_csv('../data/user.csv')
```

```python
print(len(train_data))
print(len(test_data))
```

```python
# データの結合
all_data = pd.concat([train_data, test_data], ignore_index=True)

# card_dataをall_dataに結合
all_data = pd.merge(all_data, card_data, on=['user_id', 'card_id'], how='left')

# user_dataをall_dataに結合
all_data = pd.merge(all_data, user_data, on='user_id', how='left')

all_data.tail()
```

```python
all_data.to_csv('../data/all_data.csv', index=False)
```
